## Prime Numbers

### Pure Python

```python
def is_prime(I):
    if I % 2 == 0: return False
    for i in range(3, int(I ** 0.5) + 1, 2):
        if I % i == 0: return False
    return True
```

## Fibonacci Numbers

### Recursive Algorithm

```python
def fib_rec_py1(n):
    if n < 2:
        return n
    else:
        return fib_rec_py1(n - 1) + fib_rec_py1(n - 2)
```

```python
%time fib_rec_py1(35)
```

```python
fib_rec_nb = numba.jit(fib_rec_py1)
```

```python
import numba

@numba.jit(nopython=True)
def fib_rec_nb(n):
    if n < 2:
        return n
    else:
        return fib_rec_nb(n - 1) + fib_rec_nb(n - 2)

%time fib_rec_nb(35)
```

```python
%%cython
def fib_rec_cy(int n):
    if n < 2:
        return n
    else:
        return fib_rec_cy(n - 1) + fib_rec_cy(n - 2)
```

```python
%time fib_rec_cy(35)
```

### Iterative Algorithm

```python
def fib_it_py(n):
    x, y = 0, 1
    for i in range(1, n + 1):
        x, y = y, x + y
    return x
```

```python
%time fib_it_py(80)
```

```python
fib_it_nb = numba.jit(fib_it_py)
```

```python
%time fib_it_nb(80)
```

```python
def mcs_pi_py(n):
    circle = 0
    for _ in range(n):
        x, y = random.random(), random.random()
        if (x ** 2 + y ** 2) ** 0.5 <= 1:
            circle += 1
    return (4 * circle) / n
```

```python
%time mcs_pi_py(n)
```

```python
mcs_pi_nb = numba.jit(mcs_pi_py)
```

```python
%time mcs_pi_nb(n)
```

```python
%%cython -a
import random
def mcs_pi_cy1(int n):
    cdef int i, circle = 0
    cdef float x
    cdef float y
    for i in range(n):
        x = random.random()
        y = random.random()
        if (x ** 2 + y ** 2) ** 0.5 <= 1:
            circle += 1
    return (4 * circle) / n
```

```python
%time mcs_pi_cy1(n)
```

```python
%%cython -a
from libc.stdlib cimport rand
cdef extern from 'limits.h':
    int INT_MAX
def mcs_pi_cy2(int n):
    cdef int i, circle = 0
    cdef float x, y
    for i in range(n):
        x, y = rand() / INT_MAX, rand() / INT_MAX
        if (x ** 2 + y ** 2) ** 0.5 <= 1:
            circle += 1
    return (4 * circle) / n
```

```python
%time mcs_pi_cy2(n)
```

## Binomial Trees

### Python

```python
import math
```

```python
S0 = 36.
T = 1.0
r = 0.06
sigma = 0.2
```

```python
def simulate_tree(M):
    dt = T / M
    u = math.exp(sigma * math.sqrt(dt))
    d = 1 / u
    S = np.zeros((M + 1, M + 1))
    S[0, 0] = S0
    z = 1
    for t in range(1, M + 1):
        for i in range(z):
            S[i, t] = S[i, t-1] * u
            S[i+1, t] = S[i, t-1] * d
        z += 1
    return S
```

```python
np.set_printoptions(formatter={'float':
                               lambda x: '%6.2f' % x})
```

```python
simulate_tree(4)
```

```python
%time simulate_tree(500)
```

```python
def simulate_tree_np(M):
    dt = T / M
    up = np.arange(M + 1)
    up = np.resize(up, (M + 1, M + 1))
    down = up.transpose() * 2
    S = S0 * np.exp(sigma * math.sqrt(dt) * (up - down))
    return S
```

### Cython

```python
%%cython -a
import numpy as np
cimport cython
from libc.math cimport exp, sqrt
cdef float S0 = 36.
cdef float T = 1.0
cdef float r = 0.06
cdef float sigma = 0.2
def simulate_tree_cy(int M):
    cdef int z, t, i
    cdef float dt, u, d
    cdef float[:, :] S = np.zeros((M + 1, M + 1),
                                  dtype=np.float32)
    dt = T / M
    u = exp(sigma * sqrt(dt))
    d = 1 / u
    S[0, 0] = S0
    z = 1
    for t in range(1, M + 1):
        for i in range(z):
            S[i, t] = S[i, t-1] * u
            S[i+1, t] = S[i, t-1] * d
        z += 1
    return np.array(S)
```

## Monte Carlo Simulation

### Python

```python
M = 100
I = 50000
```

```python
def mcs_simulation_py(p):
    M, I = p
    dt = T / M
    S = np.zeros((M + 1, I))
    S[0] = S0
    rn = np.random.standard_normal(S.shape)
    for t in range(1, M + 1):
        for i in range(I):
            S[t, i] = S[t-1, i] * math.exp((r - sigma ** 2 / 2) * dt +
                                         sigma * math.sqrt(dt) * rn[t, i])
    return S
```

```python
%time S = mcs_simulation_py((M, I))
```

```python
S[-1].mean()
```

```python
S0 * math.exp(r * T)
```

```python
K = 40.
```

```python
C0 = math.exp(-r * T) * np.maximum(K - S[-1], 0).mean()
```

### Cython &mdash; Sequential

```python
%%cython
import numpy as np
cimport numpy as np
cimport cython
from libc.math cimport exp, sqrt
cdef float S0 = 36.
cdef float T = 1.0
cdef float r = 0.06
cdef float sigma = 0.2
@cython.boundscheck(False)
@cython.wraparound(False)
def mcs_simulation_cy(p):
    cdef int M, I
    M, I = p
    cdef int t, i
    cdef float dt = T / M
    cdef double[:, :] S = np.zeros((M + 1, I))
    cdef double[:, :] rn = np.random.standard_normal((M + 1, I))
    S[0] = S0
    for t in range(1, M + 1):
        for i in range(I):
            S[t, i] = S[t-1, i] * exp((r - sigma ** 2 / 2) * dt +
                                         sigma * sqrt(dt) * rn[t, i])
    return np.array(S)
```

```python
%time S = mcs_simulation_cy((M, I))
```

## Recursive pandas Algorithm

### Data and Formula

```python
import pandas as pd
```

```python
sym = 'SPY'
```

```python
data = pd.DataFrame(pd.read_csv('http://hilpisch.com/tr_eikon_eod_data.csv',
                               index_col=0, parse_dates=True)[sym])
```

```python
alpha = 0.25
```

```python
data['EWMA'] = data[sym]
```

```python
%%time
for t in zip(data.index, data.index[1:]):
    data.loc[t[1], 'EWMA'] = (alpha * data.loc[t[1], sym] +
                              (1 - alpha) * data.loc[t[0], 'EWMA'])
```

```python
data.head()
```

```python
data[data.index > '2017-1-1'].plot(figsize=(10, 6));
```

### Pure Python

```python
def ewma_py(x, alpha):
    y = np.zeros_like(x)
    for i in range(len(x)):
        if i == 0:
            y[i] = x[i]
        else:
            y[i] = alpha * x[i] + (1 - alpha) * y[i - 1]
    return y
```

```python
%time data['EWMA_PY'] = ewma_py(data[sym], alpha)
```

```python
%time data['EWMA_PY'] = ewma_py(data[sym].values, alpha)
```

### Cython

```python
%%cython
import numpy as np
def ewma_cy(double[:] x, float alpha):
    cdef int i
    cdef double[:] y = np.empty_like(x)
    for i in range(len(x)):
        if i == 0:
            y[i] = x[i]
        else:
            y[i] = alpha * x[i] + (1 - alpha) * y[i - 1]
    return y
```

```python
%time data['EWMA_CY'] = ewma_cy(data[sym].values, alpha)
```
