# Python Infrastructure

## Docker

Roughly speaking, **Docker** allows you to run other operating systems, or multiple versions of the same operating system, on your machine. A Docker **conainer** contains and separates the file system, operating system, etc. from the host system (Windows, Mac OSX, Linux).

To install Docker see https://docs.docker.com/install/

In what follows, a **quick demo** will be shown based on a Docker container running **Ubuntu** and being hosted on **Mac OSX**:

    docker run -ti -p 9000:9000 -h py4fi -v /Users/yves/Temp/:/root/Temp/ ubuntu:latest /bin/bash

## Miniconda

The code that follows uses **Python 3.6**. For example, download and install **Miniconda 3.6** from https://conda.io/miniconda.html if you do not have conda already installed. Install Miniconda on your machine if required.

## Python Environment

Read about the management of environments under https://conda.io/docs/using/envs.html. For the following **no** Docker container is required.

For **Linux/Mac** you should execute the following lines on the shell to create a new environment with the desired packages (this is just a base install):

    conda create -n py4fi python=3.6
    source activate py4fi
    conda install numpy pandas matplotlib
    pip install plotly cufflinks
    conda install ipython jupyter
    jupyter notebook

On **Windows**, execute the following lines on the command prompt (this is just a base install):

    conda create -n py4fi python=3.6
    activate py4fi
    conda install numpy pandas matplotlib
    pip install plotly cufflinks
    pip install win-unicode-console
    set PYTHONIOENCODING=UTF-8
    conda install ipython jupyter
    jupyter notebook

By default, **Jupyter Notebook** shall pop up in the browser under http://localhost:8888 (or similar).
