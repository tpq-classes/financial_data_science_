## Cleaning Up the Raw Data

```python
data = raw.split('\n')
data
```

```python
data[1].replace('"', '')
```

```python
data[1].replace('"', '').split(',')
```

```python
data = [row.replace('"', '').split(',') for row in data[1:-1]]
```

```python
data
```
